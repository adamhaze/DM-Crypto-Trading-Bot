```python
def change(df,col_name_one, col_name_two):
    return (df[col_name_one] - df[col_name_two])

day_one['Gains']  = change(day_one,'Open','Close')
day_one['Gains']  = np.where(day_one['Gains']<0, 0, day_one['Gains'])
day_one['Losses'] = change(day_one,'Open','Close')
day_one['Losses'] = np.where(day_one['Losses']>=0, 0, day_one['Losses']*-1)

def get_first_avg(df, col_name, col_name_change, num_days):
    total = 0 
    iter_num = 0 
    for index, row in df.iterrows(): 
        if iter_num > num_days-2: 
            total+=row[col_name]
            df.loc[index, col_name_change] = total/num_days # adds the total to the for the first AVG
            break 
        df.loc[index, col_name_change] = np.nan # adds the NANS for first num_days vals
        total+=row[col_name]
        iter_num+= 1

day_one['Avg_Gain_14']=0
get_first_avg(day_one, 'Gains', 'Avg_Gain_14', 14)
day_one['Avg_Loss_14']=0
get_first_avg(day_one, 'Losses', 'Avg_Loss_14',14)



day_one['Avg_Gain_14'] = np.where(day_one['Avg_Gain_14'].shift(1).isnull() | day_one['Avg_Gain_14'].isnull() , day_one['Avg_Gain_14'], (( ( (day_one['Avg_Gain_14'].shift(1)) *13) + day_one['Gains'] )/14) ) 
day_one['Avg_Loss_14'] = np.where(day_one['Avg_Loss_14'].shift(1).isnull() | day_one['Avg_Loss_14'].isnull() , day_one['Avg_Loss_14'], ((day_one['Avg_Loss_14'].shift(1)*13) + day_one['Losses'] )/14 ) 

def get_rs(df):
    return df['Avg_Gain_14']/df['Avg_Loss_14']

day_one['RS'] = get_rs(day_one)
day_one.replace([np.inf, -np.inf], 0, inplace=True)
day_one['RS'] = np.where(((day_one['Avg_Gain_14']==0) | (day_one['Avg_Loss_14']==0)) , 0, day_one['RS'])

def get_rsi(df):
    return (100- (100 /(1+df['RS'])))
day_one['RSI']=get_rsi(day_one)

# day_one.to_csv('testing_2.csv')

day_one
```

```python
MasterData = MasterData.iloc[620: , :]
MasterData = MasterData.reset_index(drop=True)
del MasterData['index']
MasterData
```

```python
# temp_set = MasterData[0:10]
for i in range(len(MasterData)):
    u_timestamp = MasterData.loc[i,'Unix Timestamp']
    u_timestamp_len = int(math.log10(u_timestamp))+1
    if u_timestamp_len > 10 : 
        if u_timestamp_len ==13:
            MasterData.loc[i,'Unix Timestamp']= u_timestamp/1000
        elif u_timestamp_len ==12:
            MasterData.loc[i,'Unix Timestamp']= u_timestamp/100
        elif u_timestamp_len ==14:
            MasterData.loc[i,'Unix Timestamp']= u_timestamp/10000
        elif u_timestamp_len ==11:
            MasterData.loc[i,'Unix Timestamp']= u_timestamp/10
        elif u_timestamp_len ==15:
            MasterData.loc[i,'Unix Timestamp']= u_timestamp/100000
            
    if(i%100000==0):
        print(i)
MasterData
```

```python
# combine All Files into 1 
rootDir = '../data1'
master_fname= 'Master_Data_File2.csv'
file_dict={}
years= []

for dirName, subdirList, fileList in os.walk(rootDir):
    if (dirName == rootDir):
        print('In directory: %s' % dirName)
        for fname in fileList:
            split = fname.split('_')
            if (len(split)>3):
                print('\t Loading  '+fname)
                years.append(split[2])
                file_dict[split[2]]= fname

years.sort()
years.reverse()

data_files = []
for year in years: 
    file_path = os.path.join(rootDir, file_dict[year])
    tempdata = pd.read_csv(file_path, skiprows=1)
    data_files.append(tempdata)
    
master_data_file1 = pd.concat(data_files,ignore_index=True)
master_data_file1 = master_data_file1.sort_index(axis=0 ,ascending=False)
master_data_file1 = master_data_file1.reset_index(drop=True)

# master_data_file1.to_csv(master_fname, index=False)

print('done')
master_data_file1
```

```python
# deletes first entry to start at 0:00:00
master_data_file1 = master_data_file1.iloc[1439: , :]

master_data_file1 = master_data_file1.reset_index(drop=True)
del master_data_file1['Volume USD']
master_data_file1
```

```python
master_data_file1=master_data_file1.rename(columns={"unix": "Unix Timestamp", "date": "Date",'symbol':'Symbol','open':'Open','high':'High','low':'Low','close':'Close','Volume BTC':'Volume'})
master_data_file1
```

```python
def get_missing_days(date_one,date_two,data_file_1,data_file_2):
    missing_days = []
#     add first date to new data set 
    timestamp = date_one.replace(tzinfo=timezone.utc).timestamp()
    first_date_df  = data_file_1.loc[data_file_1['Unix Timestamp'] == timestamp]
    first_date_row = first_date_df.values.tolist()
    missing_days.append(first_date_row[0])
#     Loop Control
    difference = date_two-date_one
    loop_num = int(difference.total_seconds() / 60.0)
#     add missing dates
    for x in range(loop_num-1):
        new_date  =  date_one + datetime.timedelta(minutes=1+x)
        timestamp = new_date.replace(tzinfo=timezone.utc).timestamp()
        temp_data_row_df  = data_file_2.loc[data_file_2['Unix Timestamp'] == timestamp]
        if (len(temp_data_row_df)>0):
            new_data_row = temp_data_row_df.values.tolist()
            missing_days.append(new_data_row[0])
        else: 
#             use the last dates values and new unix and date. 
            last_item_values    = copy.copy(first_date_row[0])
            last_item_values[0] = timestamp
            last_item_values[1] = new_date.strftime('%Y-%m-%d %H:%M:%S')
            missing_days.append(last_item_values)
            
#   add last date 
    timestamp = date_two.replace(tzinfo=timezone.utc).timestamp()
    second_date_df  = data_file_1.loc[data_file_1['Unix Timestamp'] == timestamp]
    second_date_row = second_date_df.values.tolist()
    missing_days.append(second_date_row[0])
    return missing_days


last_date
fixed_data_set=[]
# temp_set = MasterData.iloc[2500000:2600000]
# temp_set = temp_set.reset_index(drop=True)
for i in range(len(MasterData)) :
    date = datetime.datetime.utcfromtimestamp(MasterData.loc[i,'Unix Timestamp'])
    if i !=0:
        difference = date-last_date
        if (difference.total_seconds() / 60.0 >1):
#             add missing items to new data set
            missing_list = get_missing_days(last_date,date,MasterData,master_data_file1)
            for x in missing_list:
                fixed_data_set.append(x)
            print('Difference: ', difference.total_seconds() / 60.0)
        else:
#         add item to new data set
            data_row = MasterData.loc[i].values.tolist()
            fixed_data_set.append(data_row)
    else: 
#         add first item to new data set
        data_row = MasterData.loc[i].values.tolist()
        fixed_data_set.append(data_row)
        
    last_date=date
    
# for x in fixed_data_set:
#     print(x)

print('done')
```

```python
df = pd.DataFrame(fixed_data_set, columns =['Unix Timestamp','Date','Symbol','Open','High','Low','Close','Volume']) 
for i in range(len(df)) :
    date = datetime.datetime.utcfromtimestamp(df.loc[i,'Unix Timestamp'])
    if i !=0:
        difference = date-last_date
        if (difference.total_seconds() / 60.0 >1):
            print(df.loc[i])
            print(i,' Difference: ', difference.total_seconds() / 60.0)
    last_date=date
```
